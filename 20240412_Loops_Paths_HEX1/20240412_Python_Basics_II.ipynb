### IF Condition

* Is referred to as branching

* The program can react differently to values of variables 

* Simple structure: if condition: block



```python
# value = 'richtig'
# if value == 'richtig':
#     print(value)

value = 'falsch'
if value == 'richtig':
    print(value)
```

### IF-ELSE Condition

* When the first IF condition is not true, execute the code block after ELSE



```python
value = 'falsch'
if value == 'richtig':
    print(value)
else:
    print(value)
```

#### ELIF Condition

The elif statement in Python is short for "else if". It is used to chain multiple conditions together in a single if statement

```python
x = 5

if x > 10:
    print("x is greater than 10")
    #löshe werte
elif x < 7:
    print("x is less than 10")
elif x < 7:
    print("x is less than 10")
elif x < 7:
    print("x is less than 10")
else:
    print("x is equal to 10")
```

### For-Loop

* Iteration over a sequence of objects!

```python
slist = ['one','two','three']

for a in slist:
    print(a)
```

#### continue:
* The continue statement is used within loops to skip the rest of the code inside the loop for the current iteration and proceed to the next iteration.




```python
slist = ['one','two','three']

for number in slist:
    if number =='two':
        continue
    print(number)
```

#### break:
* With break the entire loop is exited early

```python
slist = ['one','two','three']

for number in slist:
    if number =='two':
        break
    print(number)
```

#### enumerate:
* The enumerate() function, applied to an iterable type, returns not only the 
individual elements, but in addition a counting index

```python
slist = ['one','two','three']

for number in enumerate(slist):
    print(number)
```

#### pass:
* The pass statement is a null operation in Python. It is used when a statement is syntactically required but you don't want any action to occur.

```python
slist = ['one','two','three']

for number in slist:
    if number =='two':
        pass
    print(number)
```

### Data Input/Output

Reading CSV/TXT files with open(file,mode) :
* The open() function opens a file, and returns it as a file object.

Steps for reading a textfile in Python:

* Open a text file for reading by using the open() function.

* Read text from the file using the file read(), readline() or readlines() method.

* Close the file using the file close method close() !!

Reading text methods:

* read() → read all text from a file into a string.

* readline()→ read the text file line by line and return all the lines as strings.
* Readlines() → read all the lines of the text file and return them as a list of strings.

```python
file_txt = open('Teilnehmerliste.csv')
print(file_txt)
print(file_txt.read())
```

```python
file_txt = open('Teilnehmerliste.csv')
print(file_txt.readline())
print(file_txt.readline())
print(file_txt.readline())
```

#### Return value:

The return statement is used to terminate a function, i.e. to exit the function at this point. If necessary, the function can return a result.

### Datetime objects 

In Python, the datetime module provides classes for working with dates and times.



### Creating `datetime` Objects

You can create `datetime` objects using the `datetime.datetime` class constructor. You can specify the year, month, day, hour, minute, second, microsecond, and timezone (optional).

### Accessing Attributes

Once you have a `datetime` object, you can access its individual components (year, month, day, hour, minute, second, microsecond) using attributes.

### Current Date and Time

You can create a `datetime` object representing the current date and time using the `datetime.now()` method.

### Formatting `datetime` Objects

You can format `datetime` objects into strings using the `strftime()` method and a format string.

### Parsing Strings to `datetime` Objects

You can parse strings into `datetime` objects using the `strptime()` method and a format string.

### Arithmetic Operations

`datetime` objects support arithmetic operations such as addition and subtraction to calculate time differences.
