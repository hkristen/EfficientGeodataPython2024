# Abschlussarbeit EGM2024


## Thema: Bienenstöcke Standortfaktoren
Einfaches räumliches Modell mit ein paar Grundannahmen
- Wie weit können Bienen fliegen aus Literatur 
- negativer Faktor = Nähe zu Straße

## Vorgegebene Daten 
- 20 Bienenstöcke als Punkte (bienenstoecke_knittelfeld.geojson)
- Untersuchungsgebiet (research_area.geojson)
- Sentinel 2A Datensatz vom Untersuchungsgebiet (Aktualität, Phänologie, Wolken)
- *(Ergebnis von Aufgabe 4, damit ihr vor der Raster Basics Einheit bereits mit der Endarbeit beginnen könnt)*

## Aufgaben

### Satellitenbild analysieren (15 Punkte)
1) Layerstack des Sential 2A Datensatz mit 4 Kanälen (R-G-B-NIR) erstellen 
2) Zuschneiden des Layerstack auf Untersuchungsgebiet (research_area.geojson)
3) Mit Hilfe des Layerstack NDVI berechnen
4) Einfache Klassifizierung in Klassen "Wiese + Wald" und "Rest" mit NDVI Schwellwert   
--> alle Zwischenergebnisse sollen als eigenes Rasterfile im GeoTIFF Format gespeichert werden. Bitte auf nachvollziehbare Dateinamen achten! 

### Standortfaktoren berechnen (15 Punkte)
5) Download der Straßen (highway = motorway & primary & secondary) von OSM für die Research Area -> mit [QGIS tool QuickOSM](https://www.geodose.com/2020/08/how-to-download-osm-data-qgis.html)
6) Anteil Wiese/Wald Pixel im Umkreis je Bienenstock (=Buffer so groß wie Bienen lt. Literatur fliegen)  -> z.B. mit den zonal_stats des [rasterstats package](https://pythonhosted.org/rasterstats/)
7) Welcher Standort ist der Beste für jeweils Kategorie Wald/Blüte/Mischhonig (=meiste Pixel im Buffer)
8) Distanz von Bienenstock zur nächsten Straße berechnen (distance von Punkt zu Linie)    
--> die Ergebnisse der obigen Berechnungen sollen in das GEOJSON mit den Bienenstöcken als neue Spalte gespeichert werden. 

### BONUS: Auswertungen ( 15 Punkte)
9) Random sample : Funktion um X Bienenstöcke im Gebiet zu platzieren + Standortfaktoren automatisch bestimmen
10) Tile grid: Ein 500m x 500m Grid über Research Area anlegen. Für jede Gridzelle, die Umweltfaktoren bestimmen (= Rasterpixel zählen) 
11) HTML Karte der Ergebnisse
12) Als Programm mit Input / output schreiben    
-> zum Aufbessern eurer Note könnt ihr mit diesem Teil noch Bonuspunkte sammeln :)
