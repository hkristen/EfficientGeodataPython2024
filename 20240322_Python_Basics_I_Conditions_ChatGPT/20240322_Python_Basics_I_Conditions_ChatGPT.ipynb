### Numeric types
Numeric type objects represent number values. They are divided into three subgroups: Integers, Floating Point Numbers, Complex Numbers

**Integer** values belong to the int class. Specifically, integers represent positive or negative whole numbers without a decimal

```python
# Integers (int)
265
```

**Floating-point numbers** in Python are numbers defined with a decimal point. The class type is float.

```python
# Floating point numbers (float)
235.45
```

**Complex numbers** are often used in mathematical sciences. Python provides a class for complex numbers called complex


```python
# Complex numbers (complex)
2 + 4j
```

**Integers** have *unlimited precision*. There is no distinction between long and short numbers. The length depends on the computer memory.


```python
print(1, "is", type(1))
print(-1, "is", type(-1))
print(99, "is", type(99))
print(-99, "is", type(-99))
print(1234567890, "is", type(1234567890))
print(-1234567890, "is", type(-1234567890))
```

**Floating-point numbers** after 1.79×10308 evaluate to infinity. The smallest non-zero number is 5.0×10-324. Smaller numbers evaluate to zero:

```python
print(0.1, "is", type(0.1))
print(-1.0, "is", type(-1.0))
print(10., "is", type(10.))
print(123.4567890, "is", type(123.4567890))
```

```python
print(1.78e308, "is very large")
print(1.80e308, "is infinity")
print(5e-324, "is very small")
print(1e-324, "is zero")
```

### Set
**Sets** are used to store multiple items in a single variable. A set is a collection which is *unordered, unchangeable, and un-indexed* and cannot have two items with the same value.



```python
myset = {"apple", "banana", "cherry"}
print(myset)
print(type(myset))
```

#### Tuples

Tuples are used to store multiple items in a single variable. A tuple is a collection which is ordered and unchangeable.


```python
tuple1 = ("apple", "banana", "cherry")
tuple2 = (1, 5, 7, 9, 3)
tuple3 = (True, False, False)

print(tuple1)
print(tuple2)
print(tuple3)
```

#### Summary built-in data types 

- **List** is a collection which is ordered and changeable. Allows duplicate members.  
- **Tuple** is a collection which is ordered and unchangeable. Allows duplicate members.  
- **Set** is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.  
- **Dictionary** is a collection which is ordered (>Pyhton 3.7) and changeable. No duplicate members.

### AI Tools like ChatGPT for Python Help

AI tools like ChatGPT can be valuable for quick problem-solving and learning. However, they should be used with caution alongside other resources to enhance your Python journey.

### How to Access ChatGPT

You can access ChatGPT through its [web app](https://chat.openai.com/). The base model GPT3.5 is currently free to use, this might change in the future. 
Ofc there are a multitude of other emerging AI tools to help with coding such as [Github CoPilot](https://github.com/features/copilot) but first you need to understand the basic concepts of Python, to actually use such tools in a meaningful manner. 

### Limitations to to be aware of

While ChatGPT is a powerful tool, it's important to be aware of its limitations:

1. **Hallucinations:** ChatGPT might generate responses that appear plausible but are incorrect or nonsensical. Always critically evaluate the information.

2. **Knowledge Cutoff:** The model's knowledge is based on information available up to January 2022. Ensure you verify information for the latest updates.

3. **Not a Substitute for Learning:** While ChatGPT can provide helpful insights, it's not a substitute for understanding fundamental concepts. Use it as a supplementary tool for learning.

### Best Practices

When seeking help from AI tools like ChatGPT:

- **Be Specific:** Clearly articulate your question or problem to get more accurate responses. 
- **Verify Information:** Double-check information received from ChatGPT to ensure accuracy.
- **Explore Other Resources:** Don't rely solely on AI tools. Explore documentation, forums, and other learning materials for a comprehensive understanding.


### Steps to Use ChatGPT for Error Resolution

When encountering errors in the Python console, leverage ChatGPT efficiently with these five tips:

1. **Precise Querying:** Clearly state the error message and provide context. Specify the problematic code or describe the issue concisely.

2. **Interpretation Assistance:** If the error message is unclear, ask ChatGPT for an explanation. Seek help in understanding what the error signifies.

3. **Alternative Solutions:** Inquire about alternative approaches to solve the problem. ChatGPT can suggest different coding strategies or point out potential improvements.

4. **Learning Resources:** Request recommendations for learning resources related to the error. ChatGPT can suggest documentation, tutorials, or specific topics to deepen your understanding.

5. **Verification:** After receiving suggestions, verify the information. Cross-reference with official documentation and community forums to ensure accuracy and gain a comprehensive understanding.
